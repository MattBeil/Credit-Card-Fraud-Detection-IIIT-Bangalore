# Optimize 'OBJECT' Columns


```python

def optimize_objects(df: pd.DataFrame) -> pd.DataFrame:
    """
    This function optimizes all of the OBJECT columns
    within a Dataframe.
    """
    
    for col in df.select_dtypes(include=['object']):
        if not (type( df[col][0]==list )):
            num_unique_values = len(df[col].unique())
            num_total_values  = len(df[col])
            
            if float(num_unique_values/num_total_values) < 0.5:
                df['col'] = df['col'].astype('category')
    
    return df
```

```python

df_sess_3 = optimize_objects(df_sess_2)
df_sess_3.head()
```


# Optimize Numerical Values


```python

def optimize_floats(df: pd.DataFrame) -> pd.DataFrame:
    """
    This functions optimizes all of the FLOAT type 
    columns.
    """
    
    cols_float = df.select_dtypes(include=["float64"]).columns.tolist()
    df[cols_float] = df[cols_float].apply(pd.to_numeric, downcast='float')
    
    return df
    
```

```python

df_sess_3 = optimize_floats(df_sess_3)
df_sess_3.dtypes
```

```python

def optimize_ints(df: pd.DataFrame) -> pd.DataFrame:
    """
    This functions optimizes all of the INTEGER type 
    columns.
    """
    
    cols_int = df.select_dtypes(include=["int64"]).columns.tolist()
    df[cols_int] = df[cols_int].apply(pd.to_numeric, downcast='integer')
    
    return df
```

```python

df_sess_3 = optimize_ints(df_sess_3)
df_sess_3.dtypes
```
