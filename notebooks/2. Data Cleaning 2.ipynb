```python

def convert_age(days):
    """
    Convert a negative age in days
    to a positive age in years. 
    """
    
    return abs(days)/365
```

```python

df_sess_2["DAYS_BIRTH"].apply(lambda row: abs(row)/365)
```

```python

df_sess_2["Age"] = df_sess_2["DAYS_BIRTH"].apply(lambda row: convert_age(row))
```

```python

df_sess_2.drop(columns=["DAYS_BIRTH"], inplace=True)
```
